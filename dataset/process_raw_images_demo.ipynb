```python
img_path = "./data/raw/0.png"

plt.figure(figsize=(25, 75))
img = cv.imread("./data/raw/0.png")
plt.subplot(1, 3, 1)
plt.imshow(img, cmap="gray")
plt.title("raw")

enhanced_img = get_image(img_path, show=False)
plt.subplot(1, 3, 2)
plt.imshow(enhanced_img, cmap="gray")
plt.title("enhanced")

e = enhanced_img.copy()
chopped_img = chop(e.copy(), "0.png")
plt.subplot(1, 3, 3)
plt.imshow(chopped_img, cmap="gray")
plt.title("chopped")
```
